```python
#Read CSV
with open('C:/Users/Imtiyaz/Desktop/Inernational_student_Dataset/Origin_of_student_enrillment_by_year.csv', "r") as f:
    reader = csv.reader(f)
    csv = []
    for column in reader:
        csv.append(column)
```

```python
#convert list into dataframe
stud_enrollment = pd.DataFrame(csv)
```

```python
# Delete rows with index label   
stud_enrollment= stud_enrollment.drop(stud_enrollment.index[:4])
```

```python
print(stud_enrollment)
```

```python
#taking subset of data
stud_enrollment_Final = stud_enrollment[:42]
```

```python
stud_enrollment_Final = stud_enrollment_Final.drop([5], axis=0)
```

```python
stud_enrollment_Final = stud_enrollment_Final.drop([0], axis=1)
```

```python
# reseting index of dataframe 
stud_enrollment_Final = stud_enrollment_Final.reset_index(drop=True)
```

```python
# Droping rows from dataframe 
Final_data = stud_enrollment_Final.drop([0,17,33,34,39], axis=0)
```

```python
Final_data = Final_data.reset_index(drop=True)
```

```python
# Rename the columns 
Final_data.columns = ['Countries', '2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017']
```

```python
# fill missing values with zeros
Final_data = Final_data.applymap(lambda x: str(x.replace('..','0')))
```

```python
Final_data = Final_data.astype(str)
```
